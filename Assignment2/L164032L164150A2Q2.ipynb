```python
def mapFeature(X1, X2):
    res = np.ones(len(X1))
    for i in range(1,7):
        for j in range(0,i + 1):
            res = np.column_stack((res, (X1 ** (i-j)) * (X2 ** j)))
    return res
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
m=len(X)
X= pd.DataFrame(X)
X
```

```python
X=mapFeature(X.iloc[:, 0], X.iloc[:, 1])
X
```

```python
y = y[:, np.newaxis]
theta = np.zeros((X.shape[1], 1))
thetacopy= theta.copy()
theta.shape
```

```python
def sigmoid(z):
    return 1 / (1+np.exp(-z))
```

```python
def gradientDescent(X, y, theta, alpha, iterations,regTerm):
    costs = []  
    cost=0
    m=len(X)
    for i in range(iterations): 
        p= sigmoid(X @theta)
        gradient = (1/m)*(np.dot(X.T, (p - y)))
        gradient[1:] = gradient[1:] + ((regTerm/m) * np.array(theta[1:]))
        theta -= alpha * gradient
        cost = -np.average(y*np.log(p) + (1-y)*np.log(1-p))
        cost += ((regTerm/(2*m)) *np.sum(gradient[1:]**2))
        costs.append(cost)
    return costs, cost, theta
```

```python
costList, cost, theta = gradientDescent(X, y, thetacopy, 1.8, 40000,1)
```

```python
def plotDecisionBoundary(theta):
    
    #linspace
    #Return evenly spaced numbers over a specified interval.
    #Returns num evenly spaced samples, calculated over the interval [start, stop]
    x = np.linspace(-1, 1.5)
    y = np.linspace(-1, 1.5)
    X,Y = np.meshgrid(x,y)
    
    #flatten the X and Y arrays, Z height values as explained below
    X = X.flatten()
    Y = Y.flatten()
    Z = np.zeros((len(x) * len(y)))
    
    
    result = mapFeature(X, Y)
    Z = result.dot(theta)
    
    # reshape U, V, Z back to matrix
    X = X.reshape((len(x), len(y)))
    Y = Y.reshape((len(x), len(y)))
    Z = Z.reshape((len(x), len(y)))
    
    #X, Y : array-like, optional The coordinates of the values in Z.
    #array-like(N, M)The height values over which the contour is drawn.
    #levels : int or array-like, optional Determines the number and positions of the contour lines / regions.
    plt.contour(x,y,Z,0, colors='#7CFC00')
    plt.scatter(x=X1,y= Y1, marker="o",edgecolors='black', color='yellow')  
    plt.scatter(x=X2,y= Y2,marker="+", color='black') 
    plt.xlabel('Microchip Test1')
    plt.ylabel('Microchip Test2')
    plt.legend(labels=['good', 'faulty'])
```

```python
plotDecisionBoundary(theta)
```
